```python
# Fill missing 'Embarked' values with the most frequent value
most_frequent_embarked = titanic_train_data['Embarked'].mode()[0]
titanic_train_data['Embarked'] = titanic_train_data['Embarked'].fillna(most_frequent_embarked)

# Fill missing 'Cabin' values with 'Unknown'
titanic_train_data['Cabin'] = titanic_train_data['Cabin'].fillna('Unknown')

# One-hot encode categorical columns
titanic_train_data = pd.get_dummies(titanic_train_data, columns=['Sex', 'Embarked', 'Cabin'], drop_first=True)
```

```python
# Preprocess the test data similarly to the training data

# Drop 'Name' and 'Ticket' from test data
titnaic_test_data = titnaic_test_data.drop(['Name', 'Ticket'], axis=1, errors='ignore')

# Fill missing 'Age' values with the median from training data
titnaic_test_data['Age'] = titnaic_test_data['Age'].fillna(titanic_train_data['Age'].median())

# Fill missing 'Fare' values with the median from training data (if any)
titnaic_test_data['Fare'] = titnaic_test_data['Fare'].fillna(titanic_train_data['Fare'].median())

# Fill missing 'Embarked' values with the most frequent value from training data
# The 'Embarked' column in titanic_train_data was one-hot encoded and removed.
# We should use the 'most_frequent_embarked' variable which was previously calculated
# from the original titanic_train_data before one-hot encoding.
titnaic_test_data['Embarked'] = titnaic_test_data['Embarked'].fillna(most_frequent_embarked)

# Fill missing 'Cabin' values with 'Unknown'
titnaic_test_data['Cabin'] = titnaic_test_data['Cabin'].fillna('Unknown')

# One-hot encode categorical columns in test data
X_test = pd.get_dummies(titnaic_test_data, columns=['Sex', 'Embarked', 'Cabin'], drop_first=True)

# Align columns - crucial for consistent features between train and test
# Add missing columns in X_test that were present in X_train (fill with 0)
missing_cols_train = set(X_train.columns) - set(X_test.columns)
for c in missing_cols_train:
    X_test[c] = 0
# Drop columns in X_test that are not in X_train
missing_cols_test = set(X_test.columns) - set(X_train.columns)
for c in missing_cols_test:
    X_test = X_test.drop(c, axis=1)

# Ensure the order of columns is the same
X_test = X_test[X_train.columns]

# Drop 'PassengerId' from X_test as it's not a feature
X_test = X_test.drop('PassengerId', axis=1, errors='ignore')

display(X_test.head())
```
